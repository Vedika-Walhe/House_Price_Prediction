```python
def bhk_outlier(df):
    df_final = pd.DataFrame()
    for key, subdf in df.groupby('zipcode'):
        m = np.mean(subdf.area_per_bhk)
        s = np.std(subdf.area_per_bhk)
        new_df = subdf[(subdf.area_per_bhk > (m-s)) & (subdf.area_per_bhk < (m+s))]
        df_final = pd.concat([df_final,new_df],ignore_index=True)
    return df_final
```

```python
def price_outlier(df):
    remove_indices=[]
    for area, area_df in df.groupby('zipcode'):
        bhk_stats={}
        for bhk, bhk_df in area_df.groupby('bedrooms'):
            bhk_stats[bhk] = {
                'mean' : np.mean(bhk_df.price_per_sqftlot),
                'std' : np.std(bhk_df.price_per_sqftlot),
                'count' : bhk_df.shape[0]
            }
        for bhk,bhk_df in area_df.groupby('bedrooms'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                remove_indices = np.append(remove_indices, bhk_df[bhk_df.price_per_sqftlot < (stats['mean'])].index.values)
    return df.drop(remove_indices, axis='index')

price_outlier(df_final)
```

```python
def area_wise_price(df):
    ind = []
    for area, area_df in df.groupby('zipcode'):
        price_stat = {}
        for bhk,bhk_df in area_df.groupby('bedrooms'):
            price_stat[bhk] = {
                'mean' : np.mean(bhk_df.price),
                'std' : np.std(bhk_df.price),
                'count' : bhk_df.shape[0]
            }
        for bhk,bhk_df in area_df.groupby('bedrooms'):
            stats = price_stat.get(bhk-1)
            if stats and stats['count']>5:
                ind = np.append(ind, bhk_df[bhk_df.price < (stats['mean'])].index.values)
    return df.drop(ind, axis='index')

df_final = area_wise_price(df_final)
```

```python
X = df_final.drop(columns=['price','price_per_sqftlot'])
y = df_final['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=0)
```

```python
column_trns = make_column_transformer((OneHotEncoder(sparse=False),['zipcode']),remainder= 'passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trns, scaler, lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```
